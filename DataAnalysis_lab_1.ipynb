```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
insurance_data = pd.read_csv("insurance.csv")

# Convert the 'smoker' column to numeric for simplicity
insurance_data['smoker'] = insurance_data['smoker'].map({'yes': 1, 'no': 0})

# Filter the data where age is between 20 and 40
filtered_data = insurance_data[(insurance_data['age'] >= 20) & (insurance_data['age'] <= 40)]

# Separate smokers and non-smokers
smokers = filtered_data[filtered_data['smoker'] == 1]
non_smokers = filtered_data[filtered_data['smoker'] == 0]

# Plot the data
plt.scatter(non_smokers['age'], non_smokers['expenses'], c='blue', label='Non-Smokers', alpha=0.6)
plt.scatter(smokers['age'], smokers['expenses'], c='red', label='Smokers', alpha=0.7)

# Add title, axis labels, and legend
plt.title("Scatter Plot: Smokers (20-40 Age Group)", fontsize=16,color='red')
plt.xlabel("Age", fontsize=12)
plt.ylabel("expenses ($)", fontsize=12)
plt.legend()

# Count smokers in the age group
num_smokers = smokers.shape[0]
print(f"Number of smokers aged 20-40: {num_smokers}")

# Annotate the number of smokers in the plot
plt.text(30, max(filtered_data['expenses']) - 5000, f"Smokers: {num_smokers}", fontsize=12, color='red')

# Show grid for better readability
plt.grid(True, linestyle='--', alpha=0.5)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
insurance_data = pd.read_csv("insurance.csv")

# Convert 'smoker' column to numeric (for simplicity)
insurance_data['smoker'] = insurance_data['smoker'].map({'yes': 1, 'no': 0})

# Separate smokers and non-smokers
smokers = insurance_data[insurance_data['smoker'] == 1]
non_smokers = insurance_data[insurance_data['smoker'] == 0]

# Calculate the mean charges for smokers and non-smokers
mean_smoker_charges = smokers['expenses'].mean()
mean_non_smoker_charges = non_smokers['expenses'].mean()

# Filter the data for charges greater than the mean for each group
filtered_smokers = smokers[smokers['expenses'] > mean_smoker_charges]
filtered_non_smokers = non_smokers[non_smokers['expenses'] > mean_non_smoker_charges]

# Get the count of filtered smokers and non-smokers
smoker_count = filtered_smokers.shape[0]
non_smoker_count = filtered_non_smokers.shape[0]

# Create a bar plot to show the counts
categories = ['Smokers', 'Non-Smokers']
counts = [smoker_count, non_smoker_count]

plt.bar(categories, counts, color=['orange', 'blue'])

# Annotate the bars with the count values
for i, count in enumerate(counts):
    plt.text(i, count + 2, str(count), ha='center', fontsize=12)

# Add title and labels
plt.title("Number of Smokers and Non-Smokers with Charges Above the Mean", fontsize=16)
plt.ylabel("Number of People", fontsize=12)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
insurance_data = pd.read_csv("insurance.csv")

# Group by age and BMI, then calculate the mean expenses (charges) for each group
grouped_data = insurance_data.groupby(['age', 'bmi'])['expenses'].mean().reset_index()

# Create a pivot table for easier plotting, where 'age' and 'bmi' are the axes
pivot_data = grouped_data.pivot(index='age', columns='bmi', values='expenses')

# Plot the bar graph
pivot_data.plot(kind='bar', figsize=(12, 8))

# Customize the plot with titles and labels
plt.title("Average Insurance Charges by Age and BMI", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Average Expenses ($)", fontsize=12)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
insurance_data = pd.read_csv("insurance.csv")

# Convert 'smoker' column to numeric for easier processing
insurance_data['smoker'] = insurance_data['smoker'].map({'yes': 1, 'no': 0})

# Calculate the average charges for smokers and non-smokers
avg_smoker_charges = insurance_data[insurance_data['smoker'] == 1]['expenses'].mean()
avg_non_smoker_charges = insurance_data[insurance_data['smoker'] == 0]['expenses'].mean()

# Plot the comparison as a bar graph
categories = ['Smokers', 'Non-Smokers']
average_charges = [avg_smoker_charges, avg_non_smoker_charges]

plt.bar(categories, average_charges, color=['red', 'blue'])

# Annotate the bars with the average charges
for i, charge in enumerate(average_charges):
    plt.text(i, charge + 1000, f"${round(charge, 2)}", ha='center', fontsize=12)

# Add title and axis labels
plt.title("Average expenses for Smokers and Non-Smokers", fontsize=16)
plt.ylabel("Average Charges ($)", fontsize=12)

# Show the plot
plt.show()

# Print which group has higher charges
if avg_smoker_charges > avg_non_smoker_charges:
    print("Smokers have higher average expenses.")
else:
    print("Non-Smokers have higher average expenses.")
```
